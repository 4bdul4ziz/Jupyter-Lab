```python
#20BRS1185 Abdul Aziz A.B
A = np.array([[0, 0], [0, 3], [-2, 0]])
U, D, V = np.linalg.svd(A)
D_plus = np.zeros((A.shape[0], A.shape[1])).T
D_plus[:D.shape[0], :D.shape[0]] = np.linalg.inv(np.diag(D))
A_plus = V.T.dot(D_plus).dot(U.T)
A_plus
print(A_plus)
```

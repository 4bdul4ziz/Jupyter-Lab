```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from matplotlib import cm
import matplotlib.pyplot as plt
%matplotlib inline


#make data
delta = 0.08
x = y = np.arange(-4.0, 4.04, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z = Z1 * 2


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=cm.gist_earth,
                       linewidth=0, antialiased=False)
ax.set_zlim(0, 2.0)
ax.zaxis.set_major_locator(LinearLocator(6))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)
plt.savefig('3d_surface.jpg', dpi=120)
plt.show()

"""
Plot the surface change cmap
"""
fig = plt.figure(figsize=(9, 3))
ax1 = fig.add_subplot(131, projection='3d')
ax2 = fig.add_subplot(132, projection='3d')
ax3 = fig.add_subplot(133, projection='3d')
surf1 = ax1.plot_surface(X, Y, Z, cmap=cm.gist_earth, antialiased=False)
surf2 = ax2.plot_surface(X, Y, Z, cmap=cm.jet, antialiased=False)
surf3 = ax3.plot_surface(X, Y, Z, cmap=cm.cividis, antialiased=False)

ax1.set_title('cmap=gist_earth')
ax2.set_title('cmap=jet')
ax3.set_title('cmap=cividis')
#plt.savefig('3d_surface_cmap.jpg',dpi=150)
plt.show()

"""
Plot the surface change rcount, ccount
"""
fig = plt.figure(figsize=(9, 3))
ax1 = fig.add_subplot(131, projection='3d')
ax2 = fig.add_subplot(132, projection='3d')
ax3 = fig.add_subplot(133, projection='3d')
surf1 = ax1.plot_surface(X, Y, Z, cmap=cm.gist_earth,
                         rcount=25, ccount=25, antialiased=False)
surf2 = ax2.plot_surface(X, Y, Z, cmap=cm.gist_earth,
                         rcount=50, ccount=50, antialiased=False)
surf3 = ax3.plot_surface(X, Y, Z, cmap=cm.gist_earth,
                         rcount=101, ccount=101, antialiased=False)
ax1.set_title('rcount & ccount=25')
ax2.set_title('rcount & ccount=50')
ax3.set_title('rcount & ccount=101')
#plt.savefig('3d_surface_rcount_ccount.jpg',dpi=150)
plt.show()


"""
vmin, vmax
"""
fig = plt.figure(figsize=(9, 4))
ax1 = fig.add_subplot(131, projection='3d')
ax2 = fig.add_subplot(132, projection='3d')
ax3 = fig.add_subplot(133, projection='3d')
# Plot the surface change rcount,ccount
surf1 = ax1.plot_surface(X, Y, Z, cmap=cm.jet, antialiased=False)
surf2 = ax2.plot_surface(X, Y, Z, cmap=cm.jet, vmin=0.5, antialiased=False)
surf3 = ax3.plot_surface(X, Y, Z, cmap=cm.jet, vmax=1.5, antialiased=False)
ax1.set_title('')
ax2.set_title('vmin=0.5')
ax3.set_title('vmax=1.5')
fig.colorbar(surf1, ax=ax1, orientation='horizontal')
fig.colorbar(surf2, ax=ax2, orientation='horizontal')
fig.colorbar(surf3, ax=ax3, orientation='horizontal')
#plt.savefig('3d_surface_vmin_vmax.jpg',dpi=150)
plt.show()

"""
shade
"""
fig = plt.figure(figsize=(6, 3))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')
# Plot the surface change rcount,ccount
surf1 = ax1.plot_surface(X, Y, Z, color='lime', antialiased=False, shade=True)
surf2 = ax2.plot_surface(X, Y, Z, color='lime', antialiased=False, shade=False)
ax1.set_title('shade=True')
ax2.set_title('shade=False')
#plt.savefig('3d_surface_shade.jpg',dpi=150)
plt.show()

"""
antialiased
"""
fig = plt.figure(figsize=(6, 3))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')
# Plot the surface change rcount,ccount
surf1 = ax1.plot_surface(X, Y, Z, cmap='cividis', antialiased=False)
surf2 = ax2.plot_surface(X, Y, Z, cmap='cividis', antialiased=True)
ax1.set_title('antialiased=False')
ax2.set_title('antialiased=True')
#plt.savefig('3d_surface_antialiased.jpg',dpi=150)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-10, 10, 0.2)
y = np.arange(-10, 10, 0.2)
[X, Y] = np.meshgrid(x, y)
Z1 = np.sin(X*Y)
Z2 = np.sin(X-Y)
Z3 = (1 - X**2)*(1 - Y**2)
Z4 = np.exp(X) * np.cos(Y)
Z5 = np.sin(X) - np.sin(Y)
Z6 = (X - Y)/(1 + X**2 + Y**2)
fig, axs = plt.subplots(2, 3)
pc = axs[0, 0].contour(X, Y, Z1)
axs[0, 0].set_title('sin(xy)')
pc = axs[0, 1].contour(X, Y, Z2)
axs[0, 1].set_title('sin(x-y)')
pc = axs[0, 2].contour(X, Y, Z3)
axs[0, 2].set_title('(1 - x^2)(1-y^2')
pc = axs[1, 0].contour(X, Y, Z4)
axs[1, 0].set_title('e^x cos(y)')
pc = axs[1, 1].contour(X, Y, Z5)
axs[1, 1].set_title('sin(x) - sin(y)')
pc = axs[1, 2].contour(X, Y, Z6)
axs[1, 2].set_title('(x-y)/(1 + x^2 + y^2)')
plt.show()
```

```python
#find the LU factorization of the matrix
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
A = np.array([[1, 0], [-2, 1]])
B = np.array([[-2, 1], [-6, 4]])
C = np.array([[3, 0, 1], [6, 1, 1], [-3, 1, 0]])
D = np.array([[2, 0, 0], [0, -3, 1], [10, 12, 3]])

print(la.lu_factor(A))
print(la.lu_factor(B))
print(la.lu_factor(C))
print(la.lu_factor(D))
```

```python
#find the rank of the matrix
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la

A = np.array([[2, 3, 1], [0, -1, 2], [0, 0, 3]])
B = np.array([[-2, 2, -3], [2, 1, -6], [-1, -2, 0]])
rank = np.array([np.linalg.matrix_rank(A), np.linalg.matrix_rank(B)])
print(rank)
```

```python
#find the eigenvalues and eigenvectors of the matrix and plot them
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la


a=np.array([[2,3,1], [0,-1,2], [0,0,3]])
w,v= np.linalg.eig(a)
print("E-value:")
print(w)
print("E-vector")
print(v)
```

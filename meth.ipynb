```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-10, 10, 0.2)
y = np.arange(-10, 10, 0.2)
[X, Y] = np.meshgrid(x, y)
Z1 = np.sin(X*Y)
Z2 = np.sin(X-Y)
Z3 = (1 - X**2)*(1 - Y**2)
Z4 = np.exp(X) * np.cos(Y)
Z5 = np.sin(X) - np.sin(Y)
Z6 = (X - Y)/(1 + X**2 + Y**2)
fig, axs = plt.subplots(2, 3)
pc = axs[0, 0].contour(X, Y, Z1)
axs[0, 0].set_title('sin(xy)')
pc = axs[0, 1].contour(X, Y, Z2)
axs[0, 1].set_title('sin(x-y)')
pc = axs[0, 2].contour(X, Y, Z3)
axs[0, 2].set_title('(1 - x^2)(1-y^2')
pc = axs[1, 0].contour(X, Y, Z4)
axs[1, 0].set_title('e^x cos(y)')
pc = axs[1, 1].contour(X, Y, Z5)
axs[1, 1].set_title('sin(x) - sin(y)')
pc = axs[1, 2].contour(X, Y, Z6)
axs[1, 2].set_title('(x-y)/(1 + x^2 + y^2)')
plt.show()
```

```python
#defining the gram schmidts function
def gramSchmidt(x, y, z):
    x = x / np.linalg.norm(x)
    y = y - np.dot(x, y) * x
    y = y / np.linalg.norm(y)
    z = z - np.dot(x, z) * x - np.dot(y, z) * y
    z = z / np.linalg.norm(z)
    return x, y, z
    

x = np.array([1, 1, 0])
y = np.array([1, 2, 0])
z = np.array([0, 1, 2])
print(gramSchmidt(x, y, z))
```

```python
def angle_between(v1, v2):
    dot_product = v1.dot(v2)
    norms = np.linalg.norm(v1)
    np.linalg.norm(v2)
    return np.rad2deg(np.arccos(dot_product / norms))

v = np.array([1,1,0])
u = np.array([-0.5, 0.5, 0])
x= angle_between(v,u)
print(f"Angle between {v} and {u} is {x}")
```

```python
#least square
import numpy as np
from scipy.linalg import lstsq
import matplotlib.pyplot as plt
x = np.array([-2,-1,0,1,2])
y = np.array([1,2,3,4,5])
A = x[:,np.newaxis]**[0,1, 2]
print(A)
```

```python
p, res, mk, s = lstsq(A,y)
print(p)
print(res)
plt.plot(x,y,'o',label='original data', markersize=10)
plt.plot(x,p[1]*x+p[0],'r',label='fitted line')
plt.legend()
```

```python
import numpy as np


def pagerank(M, num_iterations: int = , d: float=0.85):
  N = M.shape[1]
  v = np.random.rand(N, 1)
  V = v/np.linalg.norm(v, 1)
  M_hat = (d*M+(1-d)/N)
  for i in range(num_iterations):
    v = M_hat@v
  return v


M = np.array([[0, 0, 0, 0, 1], [0.5, 0, 0, 0, 0], [
             0.5, 0, 0, 0, 0], [0, 1, 0.5, 0, 0], [0, 0, 0.5, 1, 0]])
v = pagerank(M, 100, 0.85)
print(v)
```
